2. **density** = apparent relative density (ARD) or coal mass per unit volume, including pores in coal; not bulk density or true relative density  
....    
  
    $ARD = TRD(1-void$_$fraction)$  
    From “Structure and Reactivity of Coal - A Survey of Selected Chinese Coals, Chapter 2 – Physical characteristics of coals”)
    
    Example calculation for Ill No 6:  
    $ARD$ = 1.32 $g/cm^3$ * (1-0.16) = 1109 $kg/m^3$  (Harding)  
    $ARD$ = 1.45 $g/cm^3$ * (1-0.16) = 1218 $kg/m^3$  (Huang et al)
    
    Example calculation for Sufco:  
    $ARD$ = 1.37 $g/cm^3$ * (1-0.03) = 1329 $kg/m^3$  (Herrin & Deming)  
  
    Values for TRD are given in the references below.
    
    **Sources of data**:  

   •	“Density measurements of Argonne premium coal samples,” Huang et al., Energy & Fuels (1995), 9, 20-24.  
       •  Example: For Ill No 6, TRD = 1.4529 $g/cm^3$.  
   •	“Determination of the true density of pulverized coal samples,” R. W. Stanton (1982).   
       •  Example: For a medium-volatile bituminous coal, TRD = 1.35 $g/cm^3$.  
   •	“The characteristics of American coals in relation to their conversion into clean energy fuels,” Spackman et al., Quarterly Technical Progress Report, July-September 1977, FE-2030-9, Pennsylvania State University.  
       •  Example: for an HVC coal, water density = 1.327 $g/cm^3$ (C=79.0% on daf basis) and 1.359 $g/m^3$ (C=78.2% on daf basis).  
   •	Stan Harding - “Using analyses from Riverview and Shay coals (Illinois basin coals) and calculating the percent carbon on a daf basis, I got %C values of 78.7% and 79.0%. Using the data from Table 1 (see below), TRD is 1.32 $g/cm^3$." Stan used He density because it penetrates into more pores than water.  
   •	For Sufco (from Upper Hiawatha Seam), Herrin & Deming list value of 1370 $kg/m^3$. Huang lists 1.31 & 1.405 $g/cm^3$ (Blind Canyon – high vol bit). Stanton lists 1.35 $g/cm^3$ (med vol bit). Spackman et al. list 1.319 $g/cm^3$ for HVC coal with 78.2% C on DAF basis, which is exactly the C content of Sufco.  
   •	“Porosity of coal and coal products,” O.P. Mahajan and P.L. Walker, Jr., DOE Report FE-2030-TR7, p. 12. (See table below).
